```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns

# Información
# Temperatura
DataT = pd.read_csv("10375 CARRIZAL T2m ERA5-Land 1m (3.94,-68.14).csv", sep=";")

# Precipitación
DataP = pd.read_csv("10375 CARRIZAL Pptn ERA5-Land 1m (3.94,-68.14).csv", sep=";")

# Etot --> Evapotranspiración
DataET = pd.read_csv("10375 CARRIZAL Etot ERA5-Land 1m (3.94,-68.14).csv", sep=";")

# Epot --> evaporación potencial
DataEP = pd.read_csv("10375 CARRIZAL Epot ERA5-Land 1m (3.94,-68.14).csv", sep=";")


# Conversión de valores str a int
DataT.astype(float)

# Conversión de valores str a int
DataP.astype(float)

# Conversión de valores str a int
DataET.astype(float)

# Conversión de valores str a int
DataEP.astype(float)
```
