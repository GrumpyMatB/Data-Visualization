# Precipitación

```python
Meses=["Ene","Feb","Mar","Abr","May","Jun","Jul","Ago","Sep","Oct","Nov","Dic"]
DataPfloat=DataP.astype(float)
Precm=DataPfloat.groupby(by="Month").mean()
fig, ax = plt.subplots(figsize=[15,9])
plt.bar(Meses, Precm["Pm [mm]"], color = 'mediumturquoise', label="Precipitación")
#for index, value in enumerate(Precm["Pm [mm]"]): 
    #plt.text(index, value, int(value), size=15)
plt.plot(Meses, NiñoPmean, "ob-", color = 'red', label="El niño", linewidth=1)
for index, value in enumerate(NiñoPmean): 
    plt.text(index, value, int(value), size=15)
plt.plot(Meses, NiñaPmean, "ob-", color = 'blue', label="La niña", linewidth=1)
for index, value in enumerate(NiñaPmean): 
    plt.text(index, value, int(value), size=15)
plt.plot(Meses, NeutraPmean, "ob-", color = 'k', label="Neutra", linewidth=1)
for index, value in enumerate(NeutraPmean): 
    plt.text(index, value, int(value), size=15)
plt.legend(loc="upper right", facecolor = "w", fontsize=20)

#plt.xlabel("Meses desde 1981 a 2022", size=20)
plt.ylabel("P [mm]", size=25)
plt.grid(axis='y')
plt.grid(axis='x')
plt.xticks(size=25)
plt.yticks(range(0, 500,50),size=25)
ax.minorticks_on()
ax.tick_params(which='major', length=10, width=2, direction='inout')
ax.tick_params(which='minor', length=5, width=2, direction='in')

#plt.title("CARRIZAL [10375]",size=15)
#plt.title("IQR temperatura", size=20)


plt.show()
```
